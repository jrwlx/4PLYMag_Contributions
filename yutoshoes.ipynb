```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cdf = pd.read_csv('yuto.csv')
```

```python
# remove leading and trailing whitespaces
cdf = cdf.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# colors
smoky = '#62566e'
bouqet = '#a2799d'
careyspink = '#d5a5b6'
bizarre = '#f1dcd8'
mauve = '#E0B0FF'
lavender = '#E6E6FA'
lily = '#C5A1AF'
londonhue = '#B09BA9'
chatelle = '#A999AA'
```

```python
# remove Pro part to get data on dunks only
dunk_df = cdf[cdf['video'] != 'AprilSkateboardsProPart']
```
